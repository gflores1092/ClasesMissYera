## **4. Transformación de datos**

Convertir tipos de datos de df1

```python
df1.dtypes
```

```python
df1 = (
    df1
    .rename(
        columns = {
            'Supplier': 'supplier',
            '% OTS': 'perc_ots', '# Items': 'items',
            '# Items Shipped': 'items_shipped', 'S/ NMV Shipped': 'nmv_shipped',
            '% CNCL Seller': 'perc_cancel_seller',
            '# Items CNCL by Seller': 'items_cancel_by_seller'
        }
    )
    .assign(
        perc_ots = lambda a: a['perc_ots'].str.replace('%', ''),
        nmv_shipped = lambda x: x['nmv_shipped'].str.replace('S/. ', '').str.replace(',',''),
        perc_cancel_seller = lambda x: x['perc_cancel_seller'].str.replace(' %','')
    )
    .assign(
        perc_ots = lambda x: np.where(x['perc_ots'] == '', 0, x['perc_ots']),
        items_shipped = lambda x: np.where(x['items_shipped'] == '', 0, x['items_shipped']),
        nmv_shipped = lambda x: np.where(x['nmv_shipped'] == '', 0, x['nmv_shipped']),
        perc_cancel_seller = lambda x: np.where(x['perc_cancel_seller'] == '', 0, x['perc_cancel_seller']),
        items_cancel_by_seller = lambda x: np.where(x['items_cancel_by_seller'] == '', 0, x['items_cancel_by_seller'])
    )
    .assign(
        perc_ots = lambda x: x['perc_ots'].astype('float'),
        items_shipped = lambda x: x['items_shipped'].astype('int'),
        nmv_shipped = lambda x: x['nmv_shipped'].astype('float'),
        perc_cancel_seller = lambda x: x['perc_cancel_seller'].astype('float'),
        items_cancel_by_seller = lambda x: x['items_cancel_by_seller'].astype('int')
    )
)
```
